```python
Q1_PPCC = Big_DS['PEOPLE_POSITIVE_CASES_COUNT'].quantile(0.25)
Q2_PPCC = Big_DS['PEOPLE_POSITIVE_CASES_COUNT'].quantile(0.75)

IQR_PPCC = Q2_PPCC - Q1_PPCC

Big_DS = Big_DS[~((Big_DS['PEOPLE_POSITIVE_CASES_COUNT'] < (Q1_PPCC - (1.5 * IQR_PPCC))) | (Big_DS['PEOPLE_POSITIVE_CASES_COUNT'] > (Q2_PPCC + (1.5 * IQR_PPCC))))]
```

```python
Q1_PDNC = Big_DS['PEOPLE_DEATH_NEW_COUNT'].quantile(0.25)
Q2_PDNC = Big_DS['PEOPLE_DEATH_NEW_COUNT'].quantile(0.75)

IQR_PDNC = Q2_PDNC - Q1_PDNC

Big_DS = Big_DS[~((Big_DS['PEOPLE_DEATH_NEW_COUNT'] < (Q1_PDNC - (1.5 * IQR_PDNC))) | (Big_DS['PEOPLE_DEATH_NEW_COUNT'] > (Q2_PDNC + (1.5 * IQR_PDNC))))]
```
